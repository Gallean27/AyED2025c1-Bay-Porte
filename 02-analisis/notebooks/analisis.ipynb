# O(n)

```python
def imprimir_items(n):
    for i in range(n):
        print(i)
        
imprimir_items(10)
```

- El parámetro de la función es n y se ejecutaron n operaciones
- La función imprimir_items(n) tiene un orden de complejidad O(n) porque el tiempo de ejecución de la función crece linealmente con el tamaño de la entrada n

# Descartar constantes

Descartar constantes es una forma de simplificar nuestra notacion O-Grande. Funciona de la siguiente manera.

```python
def imprimir_items(n):
    for i in range(n):
        print(i)

    for j in range(n):
        print(j)
        
imprimir_items(10)
```

En la ejecución del programa anterior, pasamos un 10 como parámetro y obtuvimos 20 items. 
Cada ciclo for se ejecutó con una complejidad temporal de n.
Es decir, que en total tenemos n + n = 2n y nuestro código corre con una complejidad temporal de O(2n).
Pero para estos casos descartamos las constantes y simplemente decimos que corre con una 
complejidad temporal de O(n).

# O(n^2)

```python
def imprimir_items(n):
    for i in range(n):
        for j in range(n):
            print(i, j)
        
imprimir_items(10)
```

En la ejecución del programa anterior, pasamos un 10 como parámetro y obtuvimos 100 items. Cada ciclo for ejecutó n veces otro ciclo for de n iteraciones. Es decir, que en total tenemos n * n = n^2. A eso lo llamamos O(n^2)

# O(n^3), O(n^4), etc

```python
def imprimir_items(n):
    for i in range(n):
        for j in range(n):
            for k in range(n):
                print(i, j, k)
        
imprimir_items(10)
```

# Descartar terminos NO dominantes

Esta es otra técnica de simplificación

```python
def imprimir_items(n):
    for i in range(n):
        for j in range(n):
              print(i, j)
    
    for k in range(n):
        print(k)

imprimir_items(10)
```

- Los primeros dos ciclos for se ejecutan con una complejidad temporal de O(n^2). 
- El tercer ciclo for se ejecuta con una complejidad temporal de O(n)
- En total sería O(n^2 + n)
- Pero solamente nos interesa el comportamiento del codigo cuando n es lo suficientemente grande. Llegado a ese punto, solamente nos va a interesar el aporte que haga n^2, y vamos a descartar el aporte de n.
- Es decir, que para simplificar, vamos a decir que el código anterior tiene un orden de complejidad O(n^2)

# O(log n)

```python
def imprimir_items(n):
    i = n
    cont = 1
    while i > 1:
        i = i // 2  # Reducimos el valor a la mitad en cada iteración
        print(cont, ' - ', i)        
        cont += 1

imprimir_items(160000)
```

- la operación dentro del bucle divide i por 2 en cada iteración
- el número de iteraciones es proporcional al lagoritmo en base 2 de n, es decir, O(log n)
- una pregunta para analizar este algoritmo es, ¿cuantas veces tengo que dividir el 16 para llegar a 1? Rta: 4 veces

```python
imprimir_items(1073741824)
```

# O(n * log n)

Un ejemplo típico de complejidad O(n log n) es un algoritmo que realiza n veces un proceso logarítmico en cada iteración

```python
def n_logaritmico(n):
    for i in range(n):  # Bucle lineal O(n)
        j = n
        while j > 1:  # Bucle logarítmico O(log n)
            j = j // 2
            print(f"i: {i}, j: {j}")

n_logaritmico(8)
```

# Diferentes parámetros de entrada

```python
def imprimir_items(n):
    for i in range(n):
        print(i)

    for j in range(n):
        print(j)

# Esto era O(n) + O(n) = O(2n) = O(n)

imprimir_items(10) # 20 operaciones
```

```python
def imprimir_items(a, b):
    for i in range(a):
        print(i)

    for j in range(b):
        print(j)

# Esto ES O(a) + O(b) = O(a + b)

imprimir_items(5, 10) # 15 operaciones
```

Los bucles no están anidados, por lo que las operaciones del primer bucle no afectan al segundo

```python
def imprimir_items(a, b):
    for i in range(a):
        for j in range(b):
            print(i, j)

# Esto ES O(a) * O(b) = O(a * b)

imprimir_items(5, 10) # 50 operaciones
```

Los bucles sí estan anidados, por lo que el número de operaciones es el producto de a y b
